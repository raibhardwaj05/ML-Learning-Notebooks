### The above method would take long and more effort but there is a direct method to do so

```python
df_encode = pd.get_dummies(df, drop_first= True)
df_encode = df_encode.astype(int)
df_encode
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

cols = ['Age', 'RestingBP',	'Cholesterol', 'MaxHR',	'Oldpeak']
scalar = StandardScaler()

df_encode[cols] = scalar.fit_transform(df_encode[cols]) 
df_encode.head()
```

# Feature Selection/ Extraction

```python
df_encode.columns
```

```python
numeric_cols = ['Age', 'RestingBP', 'Cholesterol', 'MaxHR', 'Oldpeak']
```

```python
from scipy.stats import pearsonr

correlations = {
    feature: pearsonr(df_encode[feature], df_encode['HeartDisease'])[0]
    for feature in numeric_cols
}

correlations_df = pd.DataFrame(list(correlations.items()), columns= ['Feature', 'Correlation'])
correlations_df.sort_values(by = 'Correlation', ascending= False)
```

```python
categorical_features = [
       'FastingBS', 'Sex_M', 'ChestPainType_ATA', 'ChestPainType_NAP',
       'ChestPainType_TA', 'RestingECG_Normal', 'RestingECG_ST',
       'ExerciseAngina_Y', 'ST_Slope_Flat', 'ST_Slope_Up']
```

```python
from scipy.stats import chi2_contingency

alpha = 0.05
chi2_result = {}

df_encode['HeartDisease_bin'] = df_encode['HeartDisease'] # categorical target doesn't need binning

for cols in categorical_features: 
    contigency = pd.crosstab(df_encode[cols], df_encode['HeartDisease_bin'])

    chi2_stat, p_val, _, _ = chi2_contingency(contigency)

    desision = 'Reject Null (keep feature)' if p_val < alpha else 'Accept Null (drop feature)'

    chi2_result[cols] = {
        'chi2_stats': chi2_stat,
        'p_val': p_val, 
        'Desision': desision
    }

chi2_df = pd.DataFrame(chi2_result).T
chi2_df = chi2_df.sort_values(by = 'p_val')
chi2_df
```

```python
final_features = [
    'Age',
    'MaxHR',
    'Oldpeak',
    'FastingBS',
    'Sex_M',
    'ChestPainType_ATA',
    'ChestPainType_NAP',
    'RestingECG_Normal',
    'RestingECG_ST',
    'ExerciseAngina_Y',
    'ST_Slope_Flat',
    'ST_Slope_Up'
]

df_final = df_encode[final_features]
```
