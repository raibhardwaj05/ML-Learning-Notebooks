```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df1 = pd.read_csv("train.csv")
df1
```

```python
df1['Order Date'] = pd.to_datetime(df1['Order Date'], dayfirst= True)
```

```python
df1['Ship Date'] = pd.to_datetime(df1['Ship Date'], dayfirst= True)
```

```python
# .dt.days extracts the number of days from a time difference
# .dt is a datetime accessor in pandas.

df1['Shipping Days'] = (df1['Ship Date'] - df1['Order Date']).dt.days
```

```python
df1.head()
```

```python
df1.info()
```

```python
df = df1.copy()
```

```python
df.head()
```

```python
df.drop(columns= ['Row ID', 'Postal Code'], inplace= True)
```

```python
df.dtypes
```

```python
df.columns
```

```python
# what is the average shipping duration
df['Shipping Days'].mean().round().astype(int)
```

```python
# how shipping duration vary across region (always per order metric)
sns.boxplot(x = 'Region', y = 'Shipping Days', data= df, palette= 'rainbow')
plt.title("Shipping Days vs Region")
```

Shipping duration is largely consistent across all regions, with similar median values and comparable variability. This suggests that regional location does not significantly impact shipping time, indicating uniform shipping performance across regions.

```python
# shipping duration vs shipping modes
sns.boxplot(x = 'Ship Mode', y = 'Shipping Days', data= df, palette= 'rainbow')
plt.title("Shipping Days vs Ship Modes")
```

Shipping duration varies significantly across shipping modes. Same Day and First Class shipping options are faster and more consistent, while Standard Class has the longest median shipping time and the highest variability, indicating greater risk of delays.

```python
# sales vs category
sales = df.groupby(by='Category')['Sales'].sum()
sales
```

```python
sns.barplot(x= sales.index, y = sales.values, palette='rainbow')
```

Technology generates the highest total sales among all product categories, followed by Furniture, while Office Supplies contribute the least to overall revenue.

```python
# orders vs category  
orders = df.groupby(by= 'Category')['Order ID'].nunique()
orders
```

```python
sns.barplot(x=orders.index, y = orders.values, palette= 'rainbow')
```

While Technology generates the highest total sales, Office Supplies show the highest demand in terms of order volume. This suggests that Technology products are higher-value items purchased less frequently, whereas Office Supplies are lower-value items purchased more frequently.

```python
# total sales by region for each product category 
region_sales = df.groupby(by= ['Region', "Category"])['Sales'].sum().reset_index()
region_sales
```

```python
pivot_sales = region_sales.pivot(index='Region', columns= 'Category', values= 'Sales')
pivot_sales
```

```python
sns.heatmap(data= pivot_sales, annot = True, fmt= '.0f', cmap= "YlGnBu") # cmap = color map = "yellow(low) -> light green -> dark blue(high)"
plt.title("total sales by region for each product category")
```

```python
# which region contributes the highest sales per category
top_region_sales = region_sales.loc[region_sales.groupby(by = 'Category')['Sales'].idxmax().sort_values(ascending= False)]
top_region_sales
```

The heatmap shows clear variation in sales across regions and product categories. The West region generates the highest sales in Furniture and Office Supplies, while the East region records the highest sales in Technology. Overall, the West region demonstrates consistently strong performance across all categories, whereas the South region contributes the lowest sales

```python
# How does demand vary across regions for each category?
region_orders = df.groupby(by= ['Region', 'Category'])['Order ID'].nunique().reset_index()
region_orders
```

```python
pivot_region_orders = region_orders.pivot(index='Region', columns= 'Category', values= 'Order ID')
pivot_region_orders
```

```python
sns.heatmap(data= pivot_region_orders, annot= True, cmap= 'YlGnBu', fmt= '.0f')
plt.title("How does demand vary across regions for each category?")
```

Demand varies significantly across regions and product categories. The West region records the highest number of orders across all categories, followed by the East region, while the South region shows the lowest demand. Across all regions, Office Supplies consistently have the highest order volume, indicating that sales performance for this category is largely driven by high demand rather than high order value.

```python
# products with low sales
product_sales = df.groupby(by= 'Product Name')['Sales'].sum().sort_values(ascending=True).reset_index().head(10)
product_sales
```

```python
# lowest sales product across region 
product_regional_sales = (df[df['Product Name'].isin(product_sales['Product Name'])]
        .groupby(by = ['Product Name', 'Region'])['Sales']
        .sum().reset_index())

product_regional_sales
```

```python
pivot_product = product_regional_sales.pivot(index= 'Product Name', columns= 'Region', values= 'Sales')
pivot_product
```

```python
sns.heatmap(data= pivot_product, annot = True, fmt = '.0f')
plt.title("Regional Sales of Low-Selling Products")
```

The analysis identifies multiple products with consistently low sales across all regions. These products show very limited sales volumes and, in many cases, are sold in only one region with minimal values. This indicates weak overall demand rather than regional performance issues, suggesting that these products may require reevaluation in terms of promotion, pricing, or potential discontinuation.

```python
# Which products generate high sales consistently regardless of region?
high_sale_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).reset_index().head(10)
high_sale_products
```

```python
region_high_sale_products = (df[df['Product Name'].isin(high_sale_products['Product Name'])]
                             .groupby(['Product Name', 'Region'])['Sales']
                             .sum().reset_index())

region_high_sale_products
```

```python
region_count = region_high_sale_products.groupby(by= 'Product Name')['Region'].nunique()
region_count
```

```python
consistent_products = region_count[region_count == 4].index

consistent_products
```

```python
region_high_sale_products['Product Name'].isin(consistent_products)
```

```python
consistent_high_sales = region_high_sale_products[region_high_sale_products['Product Name'].isin(consistent_products)]

consistent_high_sales
```

```python
pivot_high_sale_products = consistent_high_sales.pivot(index= 'Product Name', columns= 'Region', values= 'Sales')
pivot_high_sale_products
```

```python
sns.heatmap(data= pivot_high_sale_products, annot= True, fmt = '.0f', cmap= "YlGnBu")
```

The heatmap highlights a small set of products that generate consistently high sales across all regions. The Fellowes PB500 Electric Punch Binding Machine demonstrates the most balanced and strong performance across all regions, while the HON 5400 Series Task Chairs also show high sales, particularly in the Central and East regions. Overall, these products exhibit region-agnostic demand, indicating broad market acceptance rather than reliance on a single region.

```python
# Which customers purchase most frequently, and which product categories do they prefer?
# which customer purchase most frequntly 
customers = df.groupby(['Customer ID', 'Customer Name'])['Order ID'].nunique().sort_values(ascending= False).reset_index().head(5)
customers
```

```python
customer_category = (
    df[df['Customer Name'].isin(customers['Customer Name'])]
    .groupby(['Customer Name', 'Category'])['Order ID']
    .nunique()
    .reset_index(name = 'Order count')
)
customer_category
```

```python
pivot_customers = customer_category.pivot(index='Customer Name', columns= 'Category', values= 'Order count')
pivot_customers
```

```python
sns.heatmap(pivot_customers, annot= True, fmt = '.0f', cmap= 'YlGnBu')
plt.title("Category Preferences of Frequent Customers")
```

To understand customer preferences, category-level analysis was conducted for the most frequent customers. The results show that Office Supplies are consistently the most frequently purchased category across all high-frequency customers, followed by Technology, while Furniture is purchased less often. This suggests that frequent customers tend to place repeat orders for consumable items, indicating stable and recurring demand.

```python
customer_subcategory = (
    df[df['Customer Name'].isin(customers['Customer Name'])]
    .groupby(['Customer Name', 'Sub-Category'])['Order ID']
    .nunique()
    .reset_index(name = 'Order count')
)
customer_subcategory
```

```python
pivot_customers_subcategory = customer_subcategory.pivot(index='Customer Name', columns= 'Sub-Category', values= 'Order count')
pivot_customers_subcategory
```

```python
plt.figure(figsize = (16, 8))
sns.heatmap(pivot_customers_subcategory, annot= True, fmt = '.0f', cmap= 'YlGnBu')
plt.title("Sub-Category Preferences of Frequent Customers")
```

To further understand customer preferences, sub-category level analysis was conducted for the most frequent customers. The results indicate that within Office Supplies, these customers primarily purchase Binders and Chairs, suggesting repeat demand for consumable items.
