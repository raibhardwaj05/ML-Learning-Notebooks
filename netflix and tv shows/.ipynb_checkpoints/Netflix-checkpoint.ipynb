### Handling missing data (null values)

```python
df2['director'] = df2['director'].fillna('Unknown')
```

```python
df2['cast'] = df2['cast'].fillna('Not available')
```

```python
df2['country'] = df2['country'].fillna('Unknown')
```

```python
df2['date_added'] = df2['date_added'].fillna(df2['date_added'].mode()[0])
```

```python
df2['date_added'] = pd.to_datetime(df2['date_added'], errors= 'coerce')
```

```python
df2['rating'] = df2['rating'].fillna(df2['rating'].mode()[0])
```

```python
df2[df2['date_added'].isna()][['date_added']].value_counts()
```

```python
movie_duration = df2.loc[df2['type'] == 'Movie', 'duration'].mode()[0]
movie_duration
```

```python
tv_duration = df2.loc[df2['type'] == 'TV Show', 'duration'].mode()[0]
tv_duration
```

```python
df2.loc[(df2['type'] == 'TV Show') & (df2['duration'].isna()), 'duration', ] = tv_duration
```

```python
df2.loc[(df2['type'] == 'Movie') & (df2['duration'].isna()), 'duration'] = movie_duration
```
